```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load dataset
filename ='/content/drive/MyDrive/CodingDojo Datascience/Share of US Ad Spend.xlsx'
df=pd.read_excel(filename)
df.head()
```

```python
df.info()
```

```python
#Make an array of the years.
years= df.columns[2:].values
years = list(years)
years
```

```python
#select the column's that we shall use and create a new data frame
expense_by_years = df.loc[:,df.columns[1:]]
pd.DataFrame(expense_by_years)
expense_by_years
```

```python
#make the source coloumn the index in the data frame
expense_by_years.set_index('Source',inplace=True)
expense_by_years.head()
```

```python
years= list(expense_by_years.columns.values)
years
```

```python
df1=expense_by_years.loc[:,years]
df1
```

```python
df1=expense_by_years.loc[['Total TV','Total Radio','Total Newspapers','Out-of-Home','Direct Mail','Directories','Pure-Play Internet','Search','Ex-Search'],years]
df1
```

```python
df1= df1.transpose()
df1
```

```python

plt.style.use('fivethirtyeight')
df1.plot(kind='line',marker = 'o', markersize = 10)
plt.xlabel('Year')
plt.ylabel('Expense')
plt.title('Ad Spend over the Years')

#add grid
plt.grid()
plt.legend(loc = (1.02,0));
```
